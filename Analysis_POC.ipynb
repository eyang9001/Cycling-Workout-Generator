```python
from src.parsers import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(hr_data, cad_data, pwr_data, dt) = parse_fit_file('data/Spanish Needle -3.fit')
```

```python
def exp_avg(data, win_size=20):
    df = pd.DataFrame(data)
    avgs = df.ewm(adjust=True, alpha = 0.05).mean()
    return list(avgs[0])
```

```python
ratio = [pwr_data[i]/hr_data[i] for i in range(len(hr_data))]
```

```python
xs = np.arange(len(hr_data))
fig = plt.figure(figsize=(20,10))
plt.plot(xs, hr_data)
plt.plot(xs,pwr_data)
# plt.plot(xs, cad_data)
plt.plot(xs, ratio)
plt.plot(xs, exp_avg(pwr_data, 1000))
# plt.legend(["Heart Rate", "Power", "Cadence", "exp_hr"])
plt.show()
```
